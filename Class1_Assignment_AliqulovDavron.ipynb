#Function

```python
def calculate_average(numbers):
  total = sum(numbers) # all number sum
  count = len(numbers) # all number count
  return total / count
```

```python
test1 = [80, 90, 70, 85, 95] #list 1
test2 = [55, 62, 48, 71]     # list 2
print("Average 1:", calculate_average(test1)) # test 1 result
print("Average 2:", calculate_average(test2)) # test 2 result
```

#Loops with enumerate()

```python
cities = ["Tashkent", "Samarkand", "Bukhara", "Namangan", "Fergana"] # string list
# all items print
for index, city in enumerate(cities):
  print(f"Index {index}: {city}")
```

#Combine Everything

```python
import random  #to call the library random
exam_scores = [random.randint(0, 100) for _ in range(10)] #10 len random list (0, 100)
print("Scores:", exam_scores) #print list items
```

```python
avg = calculate_average(exam_scores) #use calculate function
print(f"Class average: {avg:.1f}")
```

```python
#use condition operators list function
if avg >= 60:
  print("Result: PASS")
else:
  print("Result: FAIL")
```
